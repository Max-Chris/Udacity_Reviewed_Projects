It is also important that we take account of outliers. thess are possible values that can shift our distribution towards one area insead of giving us a true picture of what the distribution is.
Below you xcan compare the distribution of our dataset when outliers have not been removed with that that has no outlier and you could see the difference. 

```python
axrr = dfselect.hist(bins =30, edgecolor ="white",layout =(1,7), figsize =[14.70, 4.27]);
plt.suptitle("Histograms\n (All Data)", y=1.05, weight ="bold")
for ax in axrr.flatten():
    ax.set_ylabel('Frequency count of All Data')
plt.tight_layout()
plt.show()


axrr1 = dfselectLim.hist(bins =30, edgecolor ="white",layout =(1,7), figsize =[14.70, 4.27]);
plt.suptitle("Histograms\n (less than 99th pecentile)", y=1.05, weight ="bold")
for ax in axrr1.flatten():
    ax.set_ylabel('Frequency count of data less than 99th pecentile')
plt.tight_layout()
plt.show()
```

We can see the effect of outliers by comparing the histogram (all data) with histograms (less than 99thth percentile).
